```python
# 함수이름: my_extpnt
# 파라미터: prmgen
#  남성(m): 1 여성(f): 0
# return res <- 남성이면 1 여성이면 0


def my_extpnt(prmgen):
    # res=0 #넣으셨음. (함수 파악하기 좋을듯)
    if prmgen=='m':
        res=1
    elif prmgen=='f':
        res=0
    return res

my_extpnt('f') # ------0
my_extpnt('m') # ------1

# 지금은 파라미터에 f나 m 하나씩 넣음.
# gen 시리즈로 넣고 싶은데 파라미터에 떼거지로 넣는 방법은 없을까
```

```python
# 내부적으로 루프 돌면서 함수에 대입하면서 메모리 공간에 쌓아둠. -> 이걸 다시 넣어줘해야됨.
# 외우는 타이핑!!
# df['gen'] : df['gen']이라는 시리즈를
# apply(lambda gen ~~ : # lambda에 gen으로 받겠다. gen=df['gen'] 이런 느낌
# apply(lambda x ~~ : # lambda에 x로 받겠다. gen=df['gen'] 이런 느낌 이것도 가능
# lambda gen: my_extpnt(gen) : lambda의 gen에 받아온 걸 my_extpnt함수에 gen으로 넣겠다.
# lambda x: my_extpnt(x) :  lambda의 x에 받아온 걸 my_extpnt함수에 x로 넣겠다.


user = pd.read_csv('./dataset/user.csv')
user['extpnt']=user['gen'].apply(lambda x: my_extpnt(x))
user
```
