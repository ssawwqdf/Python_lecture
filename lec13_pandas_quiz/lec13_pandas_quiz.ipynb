```python
##################
# 혼자 공부한 것 #
##################

# 결과값 저장은 생략

# ------ score 복제 df 만들기

scoretttt = pd.read_csv('./dataset/score.csv')
scoretttt.isna().sum()


# ------ key를 column 이름으로 하는 임시 딕셔너리 만들기(value는 na일 경우 채울 값)

tempd={'kor':9999}


# ------ fillna에 딕셔너리를 넣으면 각 column의 이름과 같은 이름의 'key'가 연결되어 해당 column의 na에 value가 채워진다.

scoretttt.fillna(tempd)


# ------ index를 column 이름으로 하는 임시 시리즈 만들기(값은 na일 경우 채울 값)

temps=pd.Series([6666], index=['eng'])


# ------ fillna에 시리즈를 넣으면 각 column의 이름과 같은 이름의 index가 연결되어 해당 column의 na에 값이 채워진다.

scoretttt.fillna(temps)


# ------ -----응용!!
# ------ 그룹함수 사용하기

scoretttt.min() #그룹함수를 사용하면 결과는 series로 나온다.
scoretttt.fillna(scoretttt.min()) # Na가 각 column의 최소값으로 채워진다.

scoretttt.min().to_dict() # dict로 해도 마찬가지! (굳이..?)
scoretttt.fillna(scoretttt.min().to_dict())
```
