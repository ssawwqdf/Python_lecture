## Group 함수
<pre>
* db : count()        min() max() sum() avg()
* df : value_counts() min() max() sum() min()...std() var()


```python
# 각 부서별 사원수 출력
# select deptno, count(1) from emp group by deptno;
# df에서도 groupby라고 한다.(단, 띄어쓰기 ㄴㄴ)

#df.groupby()
# groupby에 들어가는 조건 by, sort, dropna ...) -> by는 첫번째라 by=안 써도 됨

# DataFrameName.groupby(by=어쩌고)[컬럼이름].그룹함수
df.groupby(by='deptno')['sal'].mean()

# select deptno,        count(deptno)           from emp group by deptno;
df.groupby(by='deptno')['deptno'].value_counts()
# 일반적으로는 groupby(by='컬럼')에 있는 column은 기본적으로 인덱스로 출력해준다.
# 하지만 value_counts()할 때 출력 column 뭔지 다시 적어줘야 에러가 안 난다!!!!

```

```python
# select deptno, sum(sal), avg(sal), min(sal), max(sal) from emp group by deptno;

# df.groupby("deptno")["sal"].sum().["sal"].avg()..? 이런 문법 없다.
df.groupby("deptno")["sal"].mean()

# select deptno, sum(sal), avg(sal), min(sal), max(sal) from emp group by deptno;
# agg+list 이용하는 방법
df.groupby("deptno")["sal"].agg(['min','max','sum','mean'])
```

```python
# select deptno, sum(sal), avg(sal), min(sal), max(sal) from emp
# where deptno=10
# group by deptno
# ;

# 조건 주는 경우 데이터프레임에 조건 먼저 준 다음 groupby!!!!
# 사실 기계학습 하면서 누구만 고르는 경우가 없긴 함..
df[df['deptno']==10].groupby("deptno")[["sal"]].agg(['min','max','sum','mean']) # 얘 왜 리스트의 리스트로 들어갔지.. 오탄가

# df[df['deptno']==10].groupby("deptno")["sal"].agg(['min','max','sum','mean'])
```

## concat axis=0 == 상하 합치기 == union ( union all??)

```python
# df10과 df20 합쳐보자!!!!!!****(상하)

# df10.concat(df20) 가능한데 안 알려주겠다. 까먹어라

# pd.concat()이용 -> 리스트에 나열해서 담아라
df1020_0=pd.concat([df10,df20]) # axis=0이 디폴트. 위+아래로 합치기
df1020_0.head()
```

```python
# 위와 같이 하면 기존 인덱스 끌고 다녀서 뒤죽박죽 -> ignore_index=True 조건 넣기.
df1020_0=pd.concat([df10,df20], ignore_index=True)
```

## concat axis=1 == 좌우 합치기 == full outer join

```python
# df10과 df20 합쳐보자!!!!!!****(좌우)

# pd.concat()이용 -> 리스트에 나열해서 담아라
df1020_1=pd.concat([df10,df20], axis=1) # 좌+우로 합치기
df1020_1.head()

# 지저분하게 나온다. -> 10번 부서에 있는 사람은 20번에 없고 20번 부서에 있는 사람은 10번에 없기 때문.
# concat의 기본값은 full outer join(깔끔하게 하려면 어떤 게 같은 건지 기준 넣어서 join 걸어줘야..)
# cf. 지저분해서 오라클에서는 굳이 full outer 지원 안 한다. -> 누가 굳이 요구하면 ANSI 문법으로 해야함.
```

# MERGE() == db: join

* 파이참에서 merge의 주요 속성, 파라미터 확인

* left 데이터셋(), right 데이터셋()   : 떼거지 안 되고 왼쪽 오른쪽 두 개만 된다. / 리스트로 감싸는 거 아니다!!
* how {'left', 'right', 'outer', 'inner'} : left 조인, right 조인, fullouter 조인, inner 조인
* 키가 동일하면 : on
* 기준이 되는 키 이름이 왼쪽 오른쪽 프레임에서 다르다면 : left_on, right_on

(suffixes: 컬럼 이름 같으면 이름 다르게 주도록 설정 가능 -> 기본은 _x, _y )

```python
dept_list=[ [10, 'ACCOUNTING', 'NEW YORK'],
            [20, 'RESEARCH', 'DALLAS'],
            [30, 'SALES', 'CHICAGO'],
            [40, 'OPERATIONS', 'BOSTON'] ]

col_list=['deptno', 'dname', 'loc']
dept=pd.DataFrame(dept_list, columns=col_list)
dept

```

```python
# select * from emp e, dept d where e.deptno=d.deptno;
# select * from emp e INNER JOIN dept d on e.deptno=d.deptno;

emp=df.drop(['new_col', 'enme_job', 'comm2', 'sal+comm2'], axis=1).copy()
emp_dept = pd.merge(dfcp,dept, how="inner", on="deptno") #deptno를 이용해  inner조인 해보자
emp_dept
```

* (left, right) outer join

```python
# ORACLE :  select * from emp e, dept d where e.deptno(+)=d.deptno;
# ANSI SQL : select * from emp e RIGHT OUTER JOIN dept d ON e.deptno=d.deptno;

emp=df.drop(['new_col', 'enme_job', 'comm2', 'sal+comm2'], axis=1).copy()

# emp_dept = pd.merge(emp, dfcp, how="left", on="deptno") #deptno를 이용해  left조인 해보자
emp_dept = pd.merge(emp,dept, how="right", on="deptno") #deptno를 이용해  right조인 해보자
emp_dept
```

* left_on, right_on
