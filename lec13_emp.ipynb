### 멀티 조건
* 조건 여러 개면 반드시 괄호로 묶어라!  
* and -> &  
* or -> |  
(&, | 안 쓰고 and, or 쓰면 " The truth value of a Series is ambiguous. Use a.empty, a.bool(), a.item(), a.any() or a.all(). " 라고 에러 뜬다.)

ex. [(df['컬럼']==10) & (df['컬럼']>=1000)]

```python
# select empno, ename, sal from emp where deptno=10 and job='salesman';
# 파이썬에서는 조건 여러 개면 반드시 ()로 묶어야 한다!!!
# dataframe에서 and는 &, or는 | !!!!!!!!!

df[['empno','ename','job']][(df['deptno']==30) & (df['job']=='SALESMAN')]
```

```python
# 30번 부서이면서 직업이 SALESMAN이거나 10번부서인 사람(직업, 부서번호 출력)
# select job, deptno from emp
# where (deptno=30 and job='SALESMAN')
#     OR (deptno=10)
# ;
df[['job','deptno']][((df['deptno']==30) & (df['job']=='SALESMAN')) | (df['deptno']==10)]
#처음부터 하나하나 치면 열고 닫고 헷갈리니까 뼈대부터 만들어라!! 겉 [] -> 속 []
#데이터에 조건 많아지면 차라리 SQL 이 편하다. 서브쿼리 형태면 특히 SQL이 편함!!!(파이썬은 for문 돌려야 됨)
#현업에서 데이터는 db로 주어지기도 하고 csv로 주어지기도 하는데
#데이터 빨리 파악하려면 RDB로 일부 불러와서 보는 게 더 편하다.
```

## Group 함수

* db : count()        min() max() sum() avg()
* df : value_counts() min() max() sum() min()...std() var()


```python
# 각 부서별 사원수 출력
# select deptno, count(1) from emp group by deptno;
# df에서도 groupby라고 한다.(단, 띄어쓰기 ㄴㄴ)

#df.groupby()
# groupby에 들어가는 조건 by, sort, dropna ...) -> by는 첫번째라 by=안 써도 됨

# DataFrameName.groupby(by=어쩌고)[컬럼이름].그룹함수
df.groupby(by='deptno')['sal'].mean()
df.groupby(by='deptno')['deptno'].value_counts()

```

```python
# select deptno, sum(sal), avg(sal), min(sal), max(sal) from emp group by deptno;

# df.groupby("deptno")["sal"].sum().["sal"].avg()..? 이런 문법 없다.
df.groupby("deptno")["sal"].mean()

# select deptno, sum(sal), avg(sal), min(sal), max(sal) from emp group by deptno;
# agg+list 이용하는 방법
df.groupby("deptno")["sal"].agg(['min','max','sum','mean'])
```

```python
# select deptno, sum(sal), avg(sal), min(sal), max(sal) from emp
# where deptno=10
# group by deptno
# ;

# 조건 주는 경우 데이터프레임에 조건 먼저 준 다음 groupby
# 사실 기계학습 하면서 누구만 고르는 경우가 없긴 함..
df[df['deptno']==10].groupby("deptno")[["sal"]].agg(['min','max','sum','mean'])
```

```python
# agg(aggregate의 약자)+dict이용하는 방법(서로 다른 column에 적용하는 경우)
# 위 select문이랑은 다름.
mydict = {"sal":"max","comm":"min"}
df.groupby("deptno").agg(mydict)
```

```python
# 사실 이렇게 쓴다. 위는 sql문에 맞춰서 group by 하는 거 보여주려고
gdf=df.describe() # 이렇게 하면 df.describe()에 대한 데이터프레임 만들어줌
                  # 숫자 있는 컬럼만 해준다.(object 이런 건 거른다.)
gdf["sal"] # 일부만 꺼내올 수도 있다.

# 이건 그냥 배운 거 써먹어보기(그냥 describe하고필요한거직접봐라)
gdf.loc[["count", "mean","min","max"],['sal','comm']]
```
